## What is Employee Attrition? 

Employee attrition happens when an employee leaves an organisation, be it for voluntary or involuntary reasons an the replacement does not happen for a long period of time or sometimes never because lets face it two people can not give the same output. And attrition leads to decrease in the size the size of organisation or department as the positions are not filled when the employees leave. 
Attrition can occur throughout an entire company or just in certain departments or divisions. This typically happens when automation or new technologies replace employees.

# Exploratory Data Analysis: 

Plotting Correlation Map Between all the Numerical Values:

```python
numeric_columns = ["Age", "DailyRate", "DistanceFromHome", "Education", "EnvironmentSatisfaction", "HourlyRate", "JobInvolvement", "JobLevel", "JobSatisfaction", "MonthlyIncome", "MonthlyRate", "NumCompaniesWorked", "PercentSalaryHike", "PerformanceRating", "RelationshipSatisfaction", "StockOptionLevel", "TotalWorkingYears", "TrainingTimesLastYear", "WorkLifeBalance", "YearsAtCompany", "YearsInCurrentRole", "YearsSinceLastPromotion", "YearsWithCurrManager"]
```

```python
num_df = hr_df[numeric_columns]

corr = num_df.corr()

plt.figure(figsize=(18,12))
sns.heatmap(corr, annot= True, cmap='Blues', fmt=".2f")
plt.title("Correlation Between Numeric Values")
plt.show()
```

#### Insights: 
- Total working years and job level has positive correlation with the attrition rate, means that employees with more experience and higher job level are much more likely to leave the organisation as they have better job opportunities and higher salary than the less experienced ones. 

- Monthly rate and salary hike percentage has less or negetive relationship with attrition as employees who are paid well and given a good salary hike are much less likely to leave the organisation. 

- Job involvement and job satisfaction are highly correlated showing that employees who are involved in their job are satisfied with the job. 

- Job level is also correlated with Monthly rate, number of companies worked, and years at company, suggesting that higher-level employees are more likely to earn higher salaries, have worked at more companies in the past, and have been with the company for longer.

### 1. Attrition level based on Overtime: 

```python
overtime_att = hr_df.groupby(['Attrition', 'OverTime']).size().unstack()

# Plotting the stacked bar chart
plt.figure(figsize=(100, 20))
ax= overtime_att.plot(kind='bar', stacked=True, colormap="rainbow", subplots=False)
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.title('Attrition Level based on Overtime')

for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{height}', (x + width/2, y + height), ha= 'right', va='bottom')


plt.show()
```

#### Insights: 
- The bar graph shows that the attrition level can be high for employees who works over time which may lead to job dissatisfaction.
- This also suggests that the company maybe suffering from understuffing which is forcing the employees working overtime. Over time is also a suggesstion of mismanagement.

### 2. Attrition level based on Marital Status: 

```python
# 3. Marital Status
marital_att = hr_df.groupby(['Attrition', 'MaritalStatus']).size().unstack()

# Plotting the stacked bar chart
plt.figure(figsize=(100, 20))
ax= marital_att.plot(kind='bar', stacked=False, colormap="rainbow", subplots=False)
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.title('Attrition Level based on Marital Status')

for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{height}', (x + width/2, y + height), ha= 'right', va='bottom')


plt.show()
```

#### Insights: 
- The bar chart shows that single employees are more likely to leave their jobs to persue new career opportunities. 
- Followed by divorced employees who are more likely to leave their job in order to relocate.

### 3. Relation between Overtime and Age: 

```python
# 4. Relation between Overtime and Age
plt.figure(figsize=(4, 3))
plt.boxplot([hr_df[hr_df['OverTime'] == 'Yes']['Age'], hr_df[hr_df['OverTime'] == 'No']['Age']], labels=['Yes', 'No'])
plt.title("Relation between Overtime and Age")
plt.xlabel("OverTime")
plt.ylabel("Age")
plt.show()
```

#### Insights: 
- The box plot shows that the median number of overtime hours worked is highest for employees in the 25-34 age group, followed by the 35-44 age group. 
- Younger employees are more likely to have jobs that require overtime or they are more interested in working overtime to earn more.
- On the other hand, older employees are experienced and they require less time to work also they are likely to have families which makes it difficult for them to work overtime.

## Recommendations: 
Based on our analysis, we propose the following recommendations:

__1. Overtime Management:__

Implement measures to manage overtime and ensure employees do not become overburdened. Consider providing flexible work options or additional support for those working overtime.

__2. Marital Status Awareness:__

Understand the needs and concerns of single employees and address factors that may contribute to higher attrition in this group. This could involve offering more inclusive benefits or programs.

__3. Education and Training:__

Focus on offering opportunities for further education and skill development, especially for employees with lower education levels. This can improve job satisfaction and retention.

__4. Commute-Friendly Policies:__

Consider policies or benefits to support employees with long commutes, such as remote work options or transportation assistance.
